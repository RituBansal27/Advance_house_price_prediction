## Handling Temporal Variables (Date Time Variables) in dataset to get better information by subtracting from year of house sold!

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## Handling Numerical Variables in dataset
### Since the numerical variables are skewed we will perform log normal distribution as we seen it in EDA analysis!

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

## Handling Rare Categorical Feature in training dataset!
### We will remove categorical variables that are present less than 1% of the observations because they don't give any useful information!

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
print("Categorical Feature Count: {}".format(len(categorical_features)))
categorical_features
```

```python
# Calculating percentage of each category under the feature column. If it is less than 1% labling as Rare_var!
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(20)
```

## Encoded the object features into numeric ones!

```python
for feature in df_col:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

## Feature scaling for dataset!

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add SalePrice variables
data = pd.concat([dataset[['SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X_train, y_train)
```

## Now, Creating the checkpoint of X_train and append y_train(Sale_price) in it and will do train_test_split!

```python
data=pd.concat([X_train,y_train],axis=1)
print(data.shape)
data.head()
```

```python
X=data.drop(['SalePrice'],axis=1)
y=data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=100)
```

```python
## Importing ml libraries and randomized search area where we will tune our hyperparameters!
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from xgboost import XGBClassifier
```

```python
classifier = XGBClassifier()
```

## Applying XGBoost with default hyperparameters!

```python
base_regressor = XGBRegressor()
base_regressor.fit(X_train,y_train)
```

```python
def evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    errors = abs(predictions - y_test)
    mape = 100 * np.mean(errors / y_test)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
base_accuracy = evaluate(base_regressor, X_test, y_test)
```

## We got a good accuracy but we will tune the hyperparameters for our personal sake!

## Tuning the hyperparameters!

```python
tuned_regressor = XGBRegressor()
```

```python
booster=['gbtree','gblinear']
base_score=[0.25,0.5,0.75,1]
```

```python
## Hyper Parameter Optimization


n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
```

```python
# Set up the random search with 4-fold cross validation
random_cv = RandomizedSearchCV(estimator=tuned_regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(X_train,y_train)
```

```python
## best estimators which we will use in tuned_regressor
random_cv.best_estimator_
```

```python
tuned_regressor = XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
             importance_type='gain', interaction_constraints='',
             learning_rate=0.05, max_delta_step=0, max_depth=15,
             min_child_weight=4, missing=None, monotone_constraints='()',
             n_estimators=900, n_jobs=0, num_parallel_tree=1,
             objective='reg:squarederror', random_state=0, reg_alpha=0,
             reg_lambda=1, scale_pos_weight=1, subsample=1, tree_method='exact',
             validate_parameters=1, verbosity=None)
```

```python
tuned_regressor.fit(X_train,y_train)
```

```python
## checking modified performance of the model
new_accuracy = evaluate(tuned_regressor, X_test, y_test)
```

## Lets see the difference b/w the base accuracy and new accuracy!

```python
print('Improvement of {:0.2f}%.'.format( 100 * (new_accuracy - base_accuracy) / base_accuracy))
```

## Lets Predict the new_test dataset on our model.

```python
y_pred=tuned_regressor.predict(new_test)
y_pred
```

## Hypermeter Tuning by using RandomSearchCV

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
print(random_grid)
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = model, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
# Best parameters which we use in tuned model
rf_random.best_params_
```

```python
rf_random.best_estimator_
```

```python
best_random = RandomForestRegressor(bootstrap=False, criterion='mse', max_depth=90,
                      max_features='sqrt', max_leaf_nodes=None,
                      min_impurity_decrease=0.0, min_impurity_split=None,
                      min_samples_leaf=1, min_samples_split=5,
                      min_weight_fraction_leaf=0.0, n_estimators=800,
                      n_jobs=None, oob_score=False, random_state=None,
                      verbose=0, warm_start=False)
```

```python
## fitting the the tuned model
best_random.fit(X_train,y_train)
```

```python
## calculating model performance
random_accuracy = evaluate(best_random, X_test, y_test)
```

```python
## Calculating the difference b/w base model and the tuned ones!
print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```
